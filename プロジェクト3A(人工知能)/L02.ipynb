Ex2. 各月の平均気温

```python
import japanize_matplotlib
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("kion10y.csv",encoding="UTF-8")

#月ごとに平均を求める
g=df.groupby(["月"])["気温"]
gg=g.sum()/g.count()
print(gg)
gg.plot()
plt.xlabel("月")
plt.ylabel("平均気温")
plt.show()
```

Ex3. 気温30度越えの調査

```python
over_bool=(df["気温"]>30)
over=df[over_bool]
cnt=over.groupby(["年"])["年"].count()
print(cnt)
cnt.plot()
plt.xlabel("年")
plt.ylabel("回数")
plt.show()
```

Ex4. 気温予測

```python
from sklearn.linear_model import LinearRegression
import numpy as np
#データを学習用とテスト用に分割する
train_year=(df["年"]<=2015)
test_year=(df["年"]>=2016)
interval=6
#過去6日分と翌日の気温をセットしたデータを作成する
def make_data(data):
  x,y=[],[]
  temps=list(data["気温"])
  for i in range(len(temps)):
    if i<interval:
      continue
    y.append(temps[i])
    xa=[]
    for p in range(interval):
      d=i+p-interval
      xa.append(temps[d])
    x.append(xa)
  return (x,y)

train_x,train_y=make_data(df[train_year])
test_x,test_y=make_data(df[test_year])

#線形回帰
lr=LinearRegression(normalize=True)
lr.fit(train_x,train_y)
pre_y=lr.predict(test_x)
plt.figure()
plt.plot(test_y,c="r")
plt.plot(pre_y,c="b")
plt.show()

#演習)気温予測の評価
"""
pre_y:機械学習により予測した気温の一覧
test_y:実際の気温
diff_y:誤差
"""
diff_y=abs(pre_y-test_y)
print(diff_y)
print("average=",sum(diff_y)/len(diff_y))
print("max=",max(diff_y))
```
