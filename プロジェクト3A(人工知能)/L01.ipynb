Ex1. AND演算の機械学習

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
learn_data=[[0,0],[0,1],[1,0],[1,1]] #学習用のデータ
learn_label=[0,0,0,1] #正解ラベル

clf=LinearSVC()
clf.fit(learn_data,learn_label)#学習

test_data=[[0,0],[0,1],[1,0],[1,1]]
test_label=clf.predict(test_data)#予測
print(test_data,"の予測結果：",test_label)
print("正解率 =",accuracy_score([0,0,0,1],test_label))
```

Ex2. XOR演算の機械学習1

```python
learn_data=[[0,0],[0,1],[1,0],[1,1]]#学習用のデータ
learn_label=[0,1,1,0]#正解データ

clf=LinearSVC()
clf.fit(learn_data,learn_label)#学習

test_data=[[0,0],[0,1],[1,0],[1,1]]
test_label=clf.predict(test_data)#予測
print(test_data,"の予測結果：",test_label)
print("正解率 =",accuracy_score([0,1,1,0],test_label))#正答率が悪い！
```

Ex3. XOR演算の機械学習2 改善版

学習結果がすぐれない場合は、
・アルゴリズムの変更
・パラメータの変更
をする。

```python
from sklearn.neighbors import KNeighborsClassifier

learn_data=[[0,0],[0,1],[1,0],[1,1]]#学習用のデータ
learn_label=[0,1,1,0]#正解データ

clf=KNeighborsClassifier(n_neighbors=1)
clf.fit(learn_data,learn_label)#学習

test_data=[[0,0],[0,1],[1,0],[1,1]]
test_label=clf.predict(test_data)#予測
print(test_data,"の予測結果：",test_label)
print("正解率 =",accuracy_score([0,1,1,0],test_label))
```
