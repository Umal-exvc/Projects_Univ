## **顔認識**

Ex1. 顔認識

```python
import matplotlib.pyplot as plt
import cv2

#カスケードファイルを指定して検出器を作成
cascade_file="haarcascade_frontalface_alt.xml"
cascade=cv2.CascadeClassifier(cascade_file)

#rgb2gray
I=cv2.imread("person.jpg")
G=cv2.cvtColor(I,cv2.COLOR_BGR2GRAY)

#顔認識を実行
face_list=cascade.detectMultiScale(G,minSize=(150,150))

#結果を確認
if len(face_list)==0:
  print("Failure!")
  quit()

#認識した部分を囲む
for x,y,w,h in face_list:
  print("顔の座標=",x,y,w,h)
  red=(0,0,255)
  cv2.rectangle(I,(x,y),(x+w,y+h),red,thickness=20)

#show
cv2.imwrite("face-detect.png",I)
plt.imshow(cv2.cvtColor(I,cv2.COLOR_RGB2GRAY))
plt.show()
```
